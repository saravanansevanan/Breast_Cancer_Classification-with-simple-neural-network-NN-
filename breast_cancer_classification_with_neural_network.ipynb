# Breast Cancer Classification with simple Neural Network

```python
#importing the dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

#Data collection and pre-processing

```python
#loading the data from sklearn
breast_cancer_data = sklearn.datasets.load_breast_cancer()
breast_cancer_data
```

```python
breast_cancer_df = pd.DataFrame(breast_cancer_data.data, columns = breast_cancer_data.feature_names)
breast_cancer_df
```

```python
#adding the target column to the data frame
breast_cancer_df['label'] = breast_cancer_data.target
```

```python
breast_cancer_df
```

```python
#shape of the dataframe
breast_cancer_df.shape
```

```python
#getting some info about the data
breast_cancer_df.info()
```

```python
#checking for missing values
breast_cancer_df.isnull().sum()
```

```python
#statistical measures about the data
breast_cancer_df.describe()
```

```python
#checking the distribution of target variable
breast_cancer_df['label'].value_counts() 
```

```python
breast_cancer_df.groupby('label').mean()
```

```python
#separating the features and target
x = breast_cancer_df.drop(columns = 'label', axis = 1)
y = breast_cancer_df['label']
```

```python
x.head()
```

```python
y.head()
```

```python
#splitting the dataset into train and test data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#Standardizing the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_std = scaler.fit_transform(x_train)
x_test_std = scaler.transform(x_test)
```

Building the Neural Network

```python
#importing tensorflow and keras
!pip install tensorflow
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers for Neural Network
model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                         keras.layers.Dense(20, activation = 'relu'),
                         keras.layers.Dense(2, activation = 'sigmoid')])
```

```python
#Compiling the neural network
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics = ['accuracy'])
```

```python
#training the neural network.
history = model.fit(x_train_std, y_train, validation_split = 0.1, epochs = 10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
